```python
import scipy
import matplotlib.pyplot as plt
import numpy as np
```

```python
xdata = np.linspace(1, 10, 10)
ydata = np.sin(xdata) + 0.5*np.log(xdata)
```

```python
xlist = np.linspace(np.min(xdata), np.max(xdata), 1000)
ylin = scipy.interpolate.interp1d(xdata, ydata, kind='linear')(xlist)
```

```python
def myfunc(x):
    return x**3 + np.sin(x)
```

```python
plt.plot(xdata, ydata, '.')
plt.plot(xlist, ylin, '-')
plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(xdata, myfunc(ydata), '.')
plt.plot(xlist, myfunc(ylin), '-')
plt.xscale('log')
plt.yscale('log')
```

```python
kvec = np.array([9.999999999999999e-05, 0.00035938136638046257, 0.001291549665014884, 0.0046415888336127815, 0.01668100537200059, 0.05994842503189409, 0.21544346900318845, 0.7742636826811277, 2.782559402207126, 10.0])
delta_approx = -np.array([-6.83201801e-02, -8.82361916e-01, -1.13916211e+01, -1.46379890e+02, -1.77675265e+03, -1.45097740e+04, -6.03663501e+04, -1.46190458e+05, -2.55526079e+05, -3.74740833e+05])
```

```python
def p(delta, k):
    return delta**2*k**(-3)

logkvec = np.log10(kvec)

klist = np.logspace(np.min(np.log10(kvec)), np.max(np.log10(kvec)), 1000)
logklist = np.log10(klist)
pk_interp = scipy.interpolate.interp1d(kvec, p(delta_approx, kvec), kind='linear')(klist)

log_pk_interp = scipy.interpolate.interp1d(logkvec, p(delta_approx, kvec), kind='linear')(logklist)

loglog_pk_interp = scipy.interpolate.interp1d(logkvec, np.log10(p(delta_approx, kvec)), kind='linear')(logklist)
```

```python
plt.plot(np.log10(klist), pk_interp, 'r-')
plt.plot(np.log10(kvec), p(delta_approx, kvec), 's', ms=10)
plt.plot(logklist, log_pk_interp, 'b-')
```

```python
#plt.plot(np.log10(klist), np.log10(pk_interp), 'r-')
plt.plot(kvec,(p(delta_approx, kvec)), 's', ms=10)
plt.loglog(10**logklist, 10**loglog_pk_interp,label="hej")
#plt.loglog(10**logklist, log_pk_interp,label="hej hej")
# logklist = np.log10(klist) 
plt.legend()
```
